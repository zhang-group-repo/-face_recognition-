```python
import face_recognition
import os
import glob
import numpy as np
import pickle
```

```python
# 遍历父文件夹下的所有子文件夹
def read_(parent_folder):
    # 存储图片文件的列表
    image_file_paths = []
    for root, dirs, files in os.walk(parent_folder):
        for dir in dirs:
            # 使用glob获取每个子文件夹内的所有图片文件
            image_files = glob.glob(os.path.join(root, dir, '*.jpg'))  # 可以根据需要修改文件扩展名
            if image_files:
                image_file_paths.extend(image_files)
    return image_file_paths

def qwe(addr):
    jpn_addr = []
    jpn_face_encoding = []
    for i,j_addr in enumerate(addr):
        # 加载图像
        obama_image = face_recognition.load_image_file(j_addr)
        q = face_recognition.face_encodings(obama_image,num_jitters=20)
        if len(q) == 0:
            print(i,"未识别出人像")
            continue
        else:
            # 计算编码，增加地址
            face_encoding = q[0]
            jpn_face_encoding.append(face_encoding)
            jpn_addr.append(j_addr)
        print(f"{i}/{len(addr)}")
    return jpn_addr,jpn_face_encoding

def cout_csv(name_addr,enciding,csv_name):

    all_face_encodings = {}
    for i,_ in enumerate(name_addr):
        all_face_encodings[name_addr[i]] = enciding[i]
    # ... etc ...
    with open(csv_name, 'wb') as f:
        pickle.dump(all_face_encodings, f)

def cin_csv(csv_name):
    with open(csv_name, 'rb') as f:
        all_face_encodings = pickle.load(f)
    
    # Grab the list of names and the list of encodings
    face_names = list(all_face_encodings.keys())
    face_encodings = np.array(list(all_face_encodings.values()))
    return face_names,face_encodings
```
